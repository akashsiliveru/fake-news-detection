```python
import pandas as pd
import numpy as np
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
import string
```

```python
true=pd.read_csv('True.csv')
```

```python
true['label']=1
```

```python
true.head()
```

```python
fake=pd.read_csv('Fake.csv')
```

```python
fake['label']=0
```

```python
fake.head()
```

```python
df=pd.concat([fake,true],axis=0)
```

```python
df.head()
```

```python
df['new_text']=df['title']+' '+df['text']
```

```python
df.head()
```

```python
df.drop(columns=['subject','date','title','text'],inplace=True)
```

```python
df.sample(5)
```

```python
df=df.sample(frac=1).reset_index(drop=True)
```

```python
df
```

```python
df['new_text']=df['new_text'].str.lower()
```

```python
df.head()
```

```python
df['new_text'].unique()
```

```python
import re

def url(text):
    return re.sub(r'http\S+|www\S+', '', text)

df['new_text'] = df['new_text'].apply(url)
```

```python
df['new_text'].unique()
```

```python
def pun(text):
    return re.sub(r'[^a-z0-9\s]','',text)
```

```python
df['new_text']=df['new_text'].apply(pun)
```

```python
df['new_text'].unique()
```

```python
def xtra(text):
    return re.sub(r'\s+', ' ', text).strip()
```

```python
df['new_text']=df['new_text'].apply(xtra)
```

```python
df['new_text'].unique()
```

```python
df['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df['new_text']

y=df['label']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
```

```python
xtrain_tfidf=vectorizer.fit_transform(xtrain)


xtest_tfidf=vectorizer.transform(xtest)
```

```python
xtrain_tfidf.toarray()
```

Converting the TF-IDF sparse matrix to dense format caused a MemoryError because the matrix size required 54.6 GB RAM. Sparse representation should be retained for efficient memory usage in text classification tasks


```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()

model.fit(xtrain_tfidf,ytrain)

ypred=model.predict(xtest_tfidf)

from sklearn.metrics import accuracy_score,confusion_matrix

accuracy_score(ytest,ypred)
```

```python
confusion_matrix(ytest,ypred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'alpha': [0.01, 0.1, 0.5, 1.0, 2.0],
    'fit_prior': [True, False]
}
```

```python
grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.fit(xtrain_tfidf, ytrain)
```

```python
ypred1=grid.predict(xtest_tfidf)
```

```python
accuracy_score(ytest,ypred1)
```

```python
confusion_matrix(ytest,ypred1)
```

## Observation

A Multinomial Naive Bayes model was trained on TF-IDF text features for fake/true news classification.

### Hyperparameter Tuning

## Conclusion
Hyperparameter tuning improved the model accuracy from **94.39%** to **96.26%** and reduced both false positives and false negatives. This demonstrates that Multinomial Naive Bayes with TF-IDF is highly effective for fake news detection.
